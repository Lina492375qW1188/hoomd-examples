```python
import numpy as np
import matplotlib.pyplot as plt
import gsd.hoomd
```

```python
traj = gsd.hoomd.open('DATA/equilibriate.gsd', 'rb')
```

```python
t_arr = []
num_arr = []
for frame in traj:
    t_arr.append( frame.configuration.step )
    num_arr.append( frame.particles.N )
    
fig, ax = plt.subplots(figsize=(4,3), nrows=1, ncols=1)
ax.plot(t_arr, num_arr)
ax.set_xlabel('Steps')
ax.set_ylabel('Number of particles')
plt.show()
```
